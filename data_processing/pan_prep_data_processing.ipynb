```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import pickle
import requests
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import torch
import h5py
import ast
```

```python
pan_prep_folder = "..data/pan_prep_data/"

train_df = pd.read_csv(pan_prep_folder + "majority_training_dataset.csv")
test_df = pd.read_csv(pan_prep_folder + "majority_testing_dataset.csv")
```

```python
#Creates a csv file of all TCR and epitope sequences to feed into PLM

combined_df = pd.concat([train_df, test_df], ignore_index=True)

cleaned_TCRs = []
semicolon = ";"
for sequence in combined_df["binding_TCR"]:
    if semicolon in sequence:
        new_sequence = sequence[:-1]
        cleaned_TCRs.append(new_sequence)
    else:
        cleaned_TCRs.append(sequence)

combined_df["binding_TCR"] = cleaned_TCRs

combined_df.to_csv(pan_prep_folder + "panprep_alldata.csv")
```

```python
cleaned_TCRs = []
semicolon = ";"
for sequence in train_df["binding_TCR"]:
    if semicolon in sequence:
        new_sequence = sequence[:-1]
        cleaned_TCRs.append(new_sequence)
    else:
        cleaned_TCRs.append(sequence)

train_df["binding_TCR"] = cleaned_TCRs
```

```python
#Splitting the testing data into 5 folds

kf = KFold(n_splits = 5, shuffle = True, random_state = 0)

splits = kf.split(test_df)

test_folds = []

for fold, (train_index, test_index) in enumerate(splits):
    test_fold_df = test_df.iloc[test_index].reset_index(drop=True)
    test_folds.append(test_fold_df)
    test_fold_df.to_csv(pan_prep_folder + "PanPrepSplits/test" + str(fold) + ".csv")
```

```python
#Splitting the training data into 5 train folds and 5 validation folds

kf = KFold(n_splits = 5, shuffle = True, random_state = 0)

splits = kf.split(train_df)

train_folds = []
val_folds = []

for fold, (train_index, val_index) in enumerate(splits):
    train_fold_df = train_df.iloc[train_index].reset_index(drop=True)
    val_fold_df = train_df.iloc[val_index].reset_index(drop=True)
    
    train_folds.append(train_fold_df)
    val_folds.append(val_fold_df)
    
    train_fold_df.to_csv(pan_prep_folder + "PanPrepSplits/train" + str(fold) + ".csv")
    val_fold_df.to_csv(pan_prep_folder + "PanPrepSplits/val" + str(fold) + ".csv")
```
