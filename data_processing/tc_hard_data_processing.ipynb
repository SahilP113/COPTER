```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import pickle
import requests
from sklearn.model_selection import train_test_split
import torch
import h5py
import ast
```

```python
#Creating new train and validation splits
data_folder = "../data/tc_hard/ds.hard-splits/pep+cdr3b/train/only-sampled-negs/"
#RN task, CDR3B only
for i in range(5):
    old_train_df = pd.read_csv(data_folder + "train-" + str(i) + ".csv")
    new_train_df, val_df = train_test_split(old_train_df, test_size = 0.1, random_state = 0)
    new_train_df.to_csv(data_folder + "new_train-" + str(i) + ".csv")
    val_df.to_csv(data_folder + "val-" + str(i) + ".csv")
```

```python
#NA task, CDR3B only
data_folder = "../data/tc_hard/ds.hard-splits/pep+cdr3b/train/only-neg-assays/"

for i in range(5):
    old_train_df = pd.read_csv("train-" + str(i) + ".csv")
    new_train_df, val_df = train_test_split(old_train_df, test_size = 0.1, random_state = 0)
    new_train_df.to_csv("new_train-" + str(i) + ".csv")
    val_df.to_csv("val-" + str(i) + ".csv")
```

```python
#RN task, CDR3A+CDR3B
data_folder = "../data/tc_hard/ds.hard-splits/pep+cdr3b+cdr3a+MHC/train/only-sampled-negs/"

for i in range(5):
    old_train_df = pd.read_csv("train-" + str(i) + ".csv")
    new_train_df, val_df = train_test_split(old_train_df, test_size = 0.1, random_state = 0)
    new_train_df.to_csv("new_train-" + str(i) + ".csv")
    val_df.to_csv("val-" + str(i) + ".csv")
```

```python
#NA task, CDR3A+CDR3B
data_folder = "../data/tc_hard/ds.hard-splits/pep+cdr3b+cdr3a+MHC/train/only-neg-assays/"

for i in range(5):
    old_train_df = pd.read_csv("train-" + str(i) + ".csv")
    new_train_df, val_df = train_test_split(old_train_df, test_size = 0.1, random_state = 0)
    new_train_df.to_csv("new_train-" + str(i) + ".csv")
    val_df.to_csv("val-" + str(i) + ".csv")
```

```python
#Creating new file called ds2.csv to feed into PLM that contains the CDR3A and CDR3B sequences combined 

full_df = pd.read_csv("tc-hard/ds.csv")
full_df2 = full_df.dropna(subset=["cdr3.alpha", "mhc.seq"])

TCR_list = []

for row in range(full_df2.shape[0]):
    TCR = full_df2.iloc[row]["cdr3.alpha"] + "X" + full_df2.iloc[row]["cdr3.beta"]
    TCR_list.append(TCR)

full_df2["TCR"] = TCR_list

full_df2.to_csv("tc-hard/ds2.csv")
```
